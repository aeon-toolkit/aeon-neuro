In a comparative study, Nguyen & Artemiadis (2018) evaluated log-Euclidean, Stein, KL, and von Neumann metrics for classifying covariance descriptors of motor imagery EEG. Interestingly, they found that all these dissimilarity metrics yielded similar classification accuracy when used in a simple $k$-NN or Riemannian centroid classifier on the original manifold

In practice, Riemannian distance functions are used in two main ways for EEG classification: (1) Directly in geometric classifiers – e.g. the Minimum Distance to Mean (MDM) classifier computes the Riemannian distance of a trial’s covariance to each class’s mean covariance and picks the closest​
